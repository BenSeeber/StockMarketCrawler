```python
tickerSymbol = 'SNP'

tickerData = yf.Ticker(tickerSymbol)
tickerDf = tickerData.history(period='1d', start='2012-1-1', end='2020-1-25')

tickerDf
```

# calculate some easy statistics over the index

```python
import numpy as np
```

```python
tickerDf["HighMean50"] = tickerDf.High.rolling(50).mean()
tickerDf["HighMean100"] = tickerDf.High.rolling(100).mean()
```

apply simple signal on data

```python
tickerDf["Position"] = np.where(tickerDf["HighMean50"]>tickerDf["HighMean100"],1,0)

tickerDf[["Position"]].groupby("Position").agg("count")
```
